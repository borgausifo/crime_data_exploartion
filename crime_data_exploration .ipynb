```python
# Importing requires packages 

import pandas as pd
import numpy as np
import sqlite3 as sql
import matplotlib as pb 
```

```python
# Creating function for reading csv files 
def csv_read(put_csv):
    df_1  = pd.read_csv(put_csv)
    return df_1 
put_csv = 'HateCrimes_csv.csv'
data_1 = csv_read(put_csv)
print('Shape of this dataframe is: {}'.format(data_1.shape))
data_1.tail()

```

```python
# Getting the snapshot of the data!
data_1.describe()
```

```python
# Creating function for readin json file 

def json_read(json_file):
    df_2 = pd.read_json(json_file)
    return df_2
json_file = 'HateCrimes_json.json'
data_2 = json_read(json_file)
print('Shape of this dataframe is: {}'.format(data_2.shape))
data_2.tail()
```

```python
data_2.describe()
```

```python
# Merging both files into one DataFrame and getting the shape information of DataFrame
data_s = pd.merge(data_1, data_2, on=['State', 'Agency_name','Agency_type'], how = 'left', indicator=True)
print('data_1 dimensions: {}'.format(data_1.shape))
print('data_2 dimensions: {}'.format(data_2.shape))
print('result dimensions: {}'.format(data_s.shape))
print('-------------------------------------------------')
print('Types of data in result: {}'.format(data_s.dtypes))

data_s.head()


```

```python
# Creating a connection to sql file and fetching data and putting into pandas DataFrame

conn = sql.connect('hatecrimes.db')
c = conn.cursor()
df_4 = c.execute('SELECT * FROM crimes')
df_5 = df_4.fetchall()
df_6 = pd.DataFrame(df_5)
df_6.columns = ['State', 'Agency_type', 'Agency_name', 'Ethnicity', 'Religion', 'Sexual_Orientation']
print('Types of data in from sql file: {}'.format(df_6.dtypes))
print('SHAPE OF sql data result is : {}'.format(df_6.shape))
df_6.head()




 

        
```

```python
# Snapshot of the DataFrame
df_6.describe()
```

```python
# Adding the Ethnicity columns from two DataFrames
ethnicity = data_s['Ethnicity'] + df_6['Ethnicity']
data_s['Ethnicity'] = ethnicity




```

```python
# Getting the snapshot of the data and finding the quarter numbers 
data_s.describe()
```

```python
#Merging both of the data from sql and other merge and getting the final result
result = pd.merge(data_s, df_6, on =['Ethnicity', 'Agency_type', 'Agency_name','State'], how = 'left')
result.head()
print('Shape of the result dataframe is: {}'.format(result.shape))
```

```python
# Describing the dataframe for quick snapshot! 
result.describe()
```

```python
result.to_csv('crime_data', sep ='t')
```

### Question: Which agency type reported maximum number of crimes in 2013? 


```python
# Finding the unique variable in dataframe 
agency_type = result.Agency_name.unique()

# Grouping the data by agency type and getting the maximum reported crimes 
groupbuy_agency_type = result.groupby('Agency_type')['quarter_1', 'quarter_2', 'quarter_3', 'quarter_4'].sum()

# Getting the total of each row for finding the total 2013 crimes 
total_agency_type = groupbuy_agency_type.sum(axis=1)
total_agency_type

```

* As we can see from the table above 'Cities' reported the most crime in 2013

### Question: What percentage of the Annual crime happened in each of the four Quarters when aggregated at Agency type level?  

```python
sum_each_columns = groupbuy_agency_type[['quarter_1', 'quarter_2', 'quarter_3','quarter_4']].sum()
l = pd.DataFrame(sum_each_columns)
l.columns = ['total']
l
```

```python
# Crimes by Agency_type 
quarter_crimes_by_agency = result.groupby('Agency_type')[['quarter_1', 'quarter_2', 'quarter_3','quarter_4']].sum()
quarter_crimes_by_agency
```

```python
# Getting the percentages of each quarter 
quarter_crimes_by_agency['first_quarter_percentage'] = quarter_crimes_by_agency['quarter_1']/1249
quarter_crimes_by_agency['second_quarter_percentage'] = quarter_crimes_by_agency['quarter_2']/1657
quarter_crimes_by_agency['third_quarter_percentage'] = quarter_crimes_by_agency['quarter_3']/1676
quarter_crimes_by_agency['forth_quarter_percentage'] = quarter_crimes_by_agency['quarter_4']/1327
quarter_crimes_by_agency
```

### Question: Which are states in which crimes have been committed against most ethnic groups? (Order by the decreasing order or ethnic groups)

### Question: Find the percentage of crimes commited against disabled and non-disabled groups in each of the four quarters. 

```python
disability_group = result.groupby('Disability')['quarter_1','quarter_2','quarter_3','quarter_4'].sum()
disability_group
```
